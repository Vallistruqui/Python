````python
import numpy as np
import pandas as pd
````

````python
a1 = pd.read_csv('1er Intervalo.csv')
a2 = pd.read_csv('2do Intervalo.csv')
a3 = pd.read_csv('3er Intervalo.csv')
````

````python
from scipy.stats import zscore
from scipy import stats

z_scores = stats.zscore(a1)

abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
df1 = a1[filtered_entries]

z_scores = stats.zscore(a2)

abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
df2 = a2[filtered_entries]

z_scores = stats.zscore(a3)

abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
df3 = a3[filtered_entries]

df1.reset_index(inplace= True,drop= True)
df2.reset_index(inplace= True,drop= True)
df3.reset_index(inplace= True,drop= True)
````

````python
TOTAL_GENERAL = len(df1)+len(df2)+len(df3)

max_df1 = max(df1.error)
min_df1 = min(df1.error)
cantidad_outliers_df1 = len(a1) - len(df1)

max_df2 = max(df2.error)
min_df2 = min(df2.error)
cantidad_outliers_df2 = len(a2) - len(df2)

max_df3 = max(df3.error)
min_df3 = min(df3.error)
cantidad_outliers_df3 = len(a3) - len(df3)
````
