```python
import numpy as np
import pandas as pd
```

```python
a1 = pd.read_csv('TODO.csv')
```

```python
from scipy.stats import zscore
from scipy import stats

z_scores = stats.zscore(a1)

abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
df1 = a1[filtered_entries]

df1.reset_index(inplace= True,drop= True)
```

```python
TOTAL_GENERAL = len(df1)
max_df1 = max(df1.error)
min_df1 = min(df1.error)
cantidad_outliers_df1 = len(a1) - len(df1)
```

```python
a = df1.error.astype(int)
b = a
a = pd.DataFrame(a)



sup = []
inf = []
porcentaje_de_valores = []
validez = []
diff = []
cantidad = []

for q in range(0,20):

  sup.append(np.percentile(a,80+q, axis = 0)[0])
  
  for p in range(0,20):

    inf.append(np.percentile(a,20-p, axis = 0)[0])
    diff.append(sup[q]-inf[p])
    iv = pd.Interval(left=inf[p], right=sup[q])
  
    for y in range(0,len(b)): 
      validez.append(b[y] in iv)

    porcentaje_de_valores.append((len(b[validez])/len(b))*100)
    cantidad.append(len(b[validez]))
    validez = []

  # print('inf',len(inf),'\n',
  #          'sup', sup[q],'\n',
  #         'diff',diff,'\n',
  #         'cantidad',cantidad,'\n',
  #         'total',len(b),'\n',
  #         'porcentaje',porcentaje_de_valores,'\n',
  #         'min',min_df1,'\n',
  #         'max',max_df1,'\n',
  #         'outliers',cantidad_outliers_df1,'\n',
  #         'total_general',TOTAL_GENERAL,'\n',
  #         'peso_intervalo',(len(b)/TOTAL_GENERAL)*100,'\n')
  
  globals()['Intervalos_1_%s' % q] = pd.DataFrame({'inf': inf, 'sup': sup[q], 'diff': diff,
                             'Captados_por_Este_Intervalo': cantidad, 'Total': len(b), 'porcentaje_de_valores': porcentaje_de_valores,
                             'Min_Intervalo': min_df1, 'Max_Intervalo': max_df1, 'Cantidad de Outliers': cantidad_outliers_df1,
                             'TOTAL_GENERAL': TOTAL_GENERAL, 'Peso de este intervalo': (len(b)/TOTAL_GENERAL)*100})
  inf = []
  porcentaje_de_valores = []
  diff = []
  cantidad = []
```
